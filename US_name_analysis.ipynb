```python
#io와 pandas 모듈 import
import io
import pandas as pd
```

```python
#우선 데이터셋을 만들자
# raw_data = {"lastName": ['Banana', 'Apple', 'Orange', 'Lemon'], "boyName": ['Red', 'Blue', 'Brown', 'Gray'], "girlName": ['Pink', 'Yellow', 'Purple', 'Beige']}
# name_df = pd.DataFrame(raw_data)

name_df = pd.read_csv('/content/drive/My Drive/US_name_list.csv')
print(name_df)
```

```python
#제일 먼저 해야할 것은, 성과 이름을 각각 조합해서 모든 조합의 경우의 수를 만들어 보자.

a = 1
fullNameBoys = []
for LN_row in name_df["Last_name"]:
  if str(LN_row) == "nan": # LN_row를 str으로 지정해주지 않으면 "nan"과 같아지지 않기 때문에 반복문을 빠져나올 수 없다. nan대신 None을 사용해 봤지만, LN_row로 분해하는 도중에 nan은 null의 값을 대변하는 변수가 아닌, 하나의 값이 되어 버렸다.
    print("Lastname이 없으므로 반복 끝!")
    break
  
  b = 1
  for BN_row in name_df["Boys_name"]:
    if str(BN_row) == "nan":
      continue
    print("%s/200번째 Lastname : %s에 대한 %s/999번째 Firstname : " %(a, LN_row, b))
    print(BN_row)
    b += 1
    BFN = str(LN_row) + str(BN_row)
    fullNameBoys.append(BFN)

  a += 1
```

```python
a = 1
fullNameGirls = []
for LN_row in name_df["Last_name"]:
  if str(LN_row) == "nan": # LN_row를 str으로 지정해주지 않으면 "nan"과 같아지지 않기 때문에 반복문을 빠져나올 수 없다. nan대신 None을 사용해 봤지만, LN_row로 분해하는 도중에 nan은 null의 값을 대변하는 변수가 아닌, 하나의 값이 되어 버렸다.
    print("Lastname이 없으므로 반복 끝!")
    break
  
  b = 1
  for GN_row in name_df["Girls_name"]:
    if str(GN_row) == "nan":
      continue
    print("%s/200번째 Lastname : %s에 대한 %s/1000번째 Firstname : " %(a, LN_row, b))
    print(GN_row)
    b += 1
    GFN = str(LN_row) + str(GN_row)
    fullNameGirls.append(GFN)

  a += 1

fullName = fullNameBoys + fullNameGirls

full_name_df = pd.DataFrame(fullName, columns=['Full_Name_BG'])
print(full_name_df)
```

```python
# full_name_df["name_length"] = len(full_name_df["Full_Name_BG"])
# print(full_name_df)

fullNameLength = []
for FN_row in full_name_df["Full_Name_BG"]:
  fullNameLength.append(len(FN_row))

print(fullNameLength)

full_name_length_df = pd.DataFrame(fullNameLength, columns=['Length'])
print(full_name_length_df)

# del full_name_df["name_length"]

full_name_df["Name_Length"] = full_name_length_df["Length"]
print(full_name_df)
```

```python
full_name_df = pd.read_csv('/content/drive/My Drive/US_name_list_total_length.csv')
del full_name_df["Unnamed: 0"]
print(full_name_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(full_name_df["Name_Length"])
plt.title('name length')
plt.show()
```

```python
import numpy as np

mu = full_name_df["Name_Length"].mean()
sigma = full_name_df["Name_Length"].std() #분산은 관측값에서 평균을 뺀 값을 제곱하고, 그것을 모두 더한 후 전체 갯수로 나눠서 구한다. 즉, 차이값의 제곱의 평균이다. 표준편차는 분산의 루트를 씌운 것이므로, 각 관측값이 평균적으로, 평균에서 얼마만큼의 거리가 떨어져있는지를 설명하는 값이다. 
max = full_name_df["Name_Length"].max()
min = full_name_df["Name_Length"].min()


print(mu)
print(sigma)
print(max)
print(min)
```
